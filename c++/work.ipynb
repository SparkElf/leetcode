```python
m=int(input())
balance=list(map(int,input().split()))
values=list(map(int,input().split()))
costs=[]
for i in range(m):
    costs.append(list(map(int,input().split())))

def dfs(idx,b1,b2,b3,sumV):
    if b1<0 or b2<0 or b3<0 : return 0
    if idx==m:return sumV
    sum1=dfs(idx+1,b1-costs[idx][0],b2-costs[idx][1],b3-costs[idx][2],sumV+values[idx])
    sum2=dfs(idx+1,b1,b2,b3,sumV)
    return max(sumV,sum1,sum2)
print(dfs(0,balance[0],balance[1],balance[2],0))

```

```python
s=input().split(' ')
n,k=int(s[0]),int(s[1])
s=list(map(int,input().split(' ')))
arr=[0]*1000000001
for i in range(len(s)):
    if s[i] in arr:
        arr[s[i]]+=1
    else : arr[s[i]]=0
l,r=0,0
def getRange():
    left,right,sum,maxSum=0,0,0,0
    while right<k :
        sum+=arr[right]
        right+=1
    maxSum=sum
    while right<1000000001:
        sum=sum+arr[right]-arr[left]
        right+=1
        left+=1
        if sum>maxSum:
            maxSum=sum
            l=left
            r=right
    return maxSum
res=getRange()
for i in range(l,r):
    if arr[i]>0:arr[i]-=1
res+=getRange()
print(res)
```
